```python
merged['HD01_VD01'] = merged['HD01_VD01'].astype(int)

merged.HD01_VD01
```

```python
#By using 'contains' function from geopandas, count how many points are in each polygon

listpoly = []

for i in range (11028):
    if merged.HD01_VD01[i] == 0:
        listpoly.append(0)
        
    else:
        
        polygon = merged.geometry[i]
    
        count = 0
        for j in range (52055):
            point = Point (collisions.X[j], collisions.Y[j])
        
            if polygon.contains(point):
                count += 1
    
        listpoly.append(count)
    
listpoly
```

```python
#make a column for the count of points in each block
merged['contained'] = listpoly
```

```python
#to make sure my code is right, the total is pretty close to the number of collision from the original data
count = 0
for i in range(11028):
    count += merged.contained[i]
    
count
```
