Grouping claims by Loss Type

```python
LT = df.groupby('Loss Type')

LT.first()
```

```python
count = LT.count()["#"]

count
```

```python
total = LT.sum()["Incurred"]

total
```

```python
mean = LT.mean()["Incurred"]

mean
```

```python
for i in len (LT):
    arr_count = {}
    arr_count.append(count[i])
    
    arr_total = {}
    arr_total.append(total[i])
    
    arr_mean = {}
    arr_mean.append(mean[i])


arr = np.array ([[count],[total],[mean]])

LT_df = pd.DataFrame(arr, columns = ['count', 'total', 'mean'])

LT_df
```
