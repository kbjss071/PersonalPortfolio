```python
def sort_by_month(df):
    Jan = []
    Feb = []
    Mar = []
    Apr = []
    May = []
    Jun = []
    Jul = []
    Aug = []
    Sep = []
    Oct = []
    Nov = []
    Dec = []
    
    for i in range (len(df['Loss Date'])):
        if df['Loss Date'][i].year == 2020:
            if df['Loss Date'][i].month == 1:
                Jan.append(i)
            elif df['Loss Date'][i].month == 2:
                Feb.append(i)
            elif df['Loss Date'][i].month == 3:
                Mar.append(i)
            elif df['Loss Date'][i].month == 4:
                Apr.append(i)
            elif df['Loss Date'][i].month == 5:
                May.append(i)
            elif df['Loss Date'][i].month == 6:
                Jun.append(i)
            elif df['Loss Date'][i].month == 7:
                Jul.append(i)
            elif df['Loss Date'][i].month == 8:
                Aug.append(i)
            elif df['Loss Date'][i].month == 9:
                Sep.append(i)
            elif df['Loss Date'][i].month == 10:
                Oct.append(i)
            elif df['Loss Date'][i].month == 11:
                Nov.append(i)
            elif df['Loss Date'][i].month == 12:
                Dec.append(i)
                
    l = [Jan, Feb, Mar, Apr, May, Jun, Jul, Aug, Sep, Oct, Nov, Dec]

    
        
    return l
```

```python
#Implemented in the df_by_month function

def dfGen (df, l):    
    
    data = []    
    for i in l:
        data.append(df.loc[i])
        
    df2 = pd.DataFrame(data)
        
    return df2
```

```python
def df_by_month (df, type_l):
    
        
    
    df_Jan = dfGen (df, type_l[0])
    df_Feb = dfGen (df, type_l[1])
    df_Mar = dfGen (df, type_l[2])
    df_Apr = dfGen (df, type_l[3])
    df_May = dfGen (df, type_l[4])
    df_Jun = dfGen (df, type_l[5])
    df_Jul = dfGen (df, type_l[6])
    df_Aug = dfGen (df, type_l[7])
    df_Sep = dfGen (df, type_l[8])
    df_Oct = dfGen (df, type_l[9])
    df_Nov = dfGen (df, type_l[10])
    df_Dec = dfGen (df, type_l[11])
    
    
    l = [df_Jan, df_Feb, df_Mar, df_Apr, df_May, df_Jun, df_Jul, df_Aug, df_Sep, df_Oct, df_Nov, df_Dec]
    
    
    return l
```

```python
def summ(df, index):
    
    if df_by_month (df, sort_by_month(df))[index].empty == True:
        return -1
    
    else:
        return sum(df_by_month (df, sort_by_month(df))[index]['This month incurred total for Summary risk(total reserve + paid)'])
```

```python
summ(liab_df, 2)
```

```python
def diffCal (df, index, type_int, case_int):
    Preset_2020 = [42000, 65000, 12000, 5500, 6000]

    Preset_2021_included = [51000, 140000, 12000, 7100, 7600]

    Preset_2021_excluded = [31000, 100000, 7100, 5500, 4000]
    
    if case_int == 0:
        diff = summ(df, index) -Preset_2020[type_int]*len(df_by_month (df, sort_by_month(df))[index])
        
    elif case_int == 1:
        diff = summ(df, index) -Preset_2021_included[type_int]*len(df_by_month (df, sort_by_month(df))[index])
        
    elif case_int == 2:
        diff = summ(df, index) -Preset_2021_excluded[type_int]*len(df_by_month (df, sort_by_month(df))[index])
    
    
    return diff
```

```python
diffCal(liab_df, 0, 0, 0)
```

```python
def monCal(df, type_int):
    Preset_2020 = [42000, 65000, 12000, 5500, 6000]

    Preset_2021_included = [51000, 140000, 12000, 7100, 7600]

    Preset_2021_excluded = [31000, 100000, 7100, 5500, 4000]

    dat = df_by_month(df, sort_by_month(df))

    type_20 = []
    type_21_included = []
    type_21_excluded = []
    case = []
    
    
    for i in range (12):
        type_20.append(diffCal(df, i, type_int, 0))
        
        type_21_included.append(diffCal(df, i, type_int, 1))        
        
        type_21_excluded.append(diffCal(df, i, type_int, 2))
        
        case.append(len(dat[i]))
        
    
    newDF = pd.DataFrame(list(zip(type_20, type_21_included, type_21_excluded, case)), columns = ['Preset_2020', 'Preset_2021_included', 'Preset_2021_excluded', '#'], index = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
        
    return newDF
```
