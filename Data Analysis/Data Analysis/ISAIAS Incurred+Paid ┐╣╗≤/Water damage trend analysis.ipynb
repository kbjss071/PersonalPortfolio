```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

water_df = pd.read_excel('Water Incurred.xlsx', header = 1)

water_df
```

```python
def comparison (df, lapse ,col1, col2):
    index = int
    
    if df.loc[lapse][col1] == None:
        index = 0
    
    elif df.loc[lapse][col1] < df.loc[lapse][col2]:
        index = 1
        
    elif df.loc[lapse][col1] == df.loc[lapse][col2]:
        index = 0
    
    elif df.loc[lapse][col1] > df.loc[lapse][col2]:
        index = -1
        
    return index
```

```python
water = water_df.replace({np.nan: None})

water
```

```python
l = []

for i in range (len(water)):
    
    l.append (0)
    
water["index"] = l

water
```

```python
for i in range (len (water)):
    
    index = comparison(water, i, "Incurred 18/1", "Incurred 18/2")
    
    prev = water.loc[i]["index"]
    
    new = prev + index
    
    water.at[i, "index"] = new
    
for i in range (len (water)):
    
    index = comparison(water, i, "Incurred 18/2", "Incurred 18/3")
    
    prev = water.loc[i]["index"]
    
    new = prev + index
    
    water.at[i, "index"] = new
    
for i in range (len (water)):
    
    index = comparison(water, i, "Incurred 18/3", "Incurred 18/4")
    
    prev = water.loc[i]["index"]
    
    new = prev + index
    
    water.at[i, "index"] = new
    
for i in range (len (water)):
    
    index = comparison(water, i, "Incurred 18/4", "Incurred 18/5")
    
    prev = water.loc[i]["index"]
    
    new = prev + index
    
    water.at[i, "index"] = new
    
for i in range (len (water)):
    
    index = comparison(water, i, "Incurred 18/5", "Incurred 18/6")
    
    prev = water.loc[i]["index"]
    
    new = prev + index
    
    water.at[i, "index"] = new
    
for i in range (len (water)):
    
    index = comparison(water, i, "Incurred 18/6", "Incurred 18/7")
    
    prev = water.loc[i]["index"]
    
    new = prev + index
    
    water.at[i, "index"] = new
    
for i in range (len (water)):
    
    index = comparison(water, i, "Incurred 18/7", "Incurred 18/8")
    
    prev = water.loc[i]["index"]
    
    new = prev + index
    
    water.at[i, "index"] = new

water
```
