```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

wind_df = pd.read_excel('Wind Incurred2.xlsx')

wind_df
```

```python
def comparison (df, lapse ,col1, col2):
    index = int
    
    if df.loc[lapse][col1] == None or df.loc[lapse][col2] == None:
        index = 0
    
    elif df.loc[lapse][col1] < df.loc[lapse][col2]:
        index = 1
        
    elif df.loc[lapse][col1] == df.loc[lapse][col2]:
        index = 0
    
    elif df.loc[lapse][col1] > df.loc[lapse][col2]:
        index = -1
        
    return index
```

```python
wind = wind_df.replace({np.nan: None})

wind
```

```python
l = []

for i in range (len(wind)):
    
    l.append (0)
    
wind["index"] = l

wind
```
